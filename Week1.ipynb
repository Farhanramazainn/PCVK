Langkah 1

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

Langkah 2

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
        "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://placekitten.com/800/571"]
# baca dan tampilkan image
# loop pada tiap url image, beberapa image dapat disimpan pada list
for url in urls:
  image = io.imread(url)                          # read image
  image = cv.resize(image, (0,0), fx=0.5, fy=0.5) # resize image to half size
  image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)  # convert color to RGB
  final_frame = cv.hconcat((image, image_2))      # concatenate image
  cv2_imshow(final_frame)                         # show image
  print('\n')
```

Langkah 3

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi," x ",lebar)
cv2_imshow(image_2)
```

Langkah 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#membuat garis horizontal ditengah image
for y in range (lebar):
  image_3[int((tinggi)/2),y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

Dalam gambar dengan resolusi tinggi, setiap area atau objek dalam gambar diwakili oleh lebih banyak pixel, sehingga menghasilkan gambar yang lebih tajam dan lebih mendekati realitas.
```python
# Garis Vertical
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in  range (tinggi):
  image_3[y, int((lebar)/2)] = [0,0,0]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)

#Garis Diagonal
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

#Garis Menyilang diagonal dari image keluaran
cv.line(image_3, (0,0), (lebar, tinggi), (0,0,0))

cv.line(image_3, (400,0), (0, tinggi), (0,0,0))

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
# Gambarlah garis putih horizontal di tengah gambar dengan panjang tertentu
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range (100,200):
  image_3[int(image_3.shape[0] / 2), y] = [255,255,255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
#Gambarlah sebuah kotak dari kumpulan piksel putih pada area tertentu pada gambar
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Menentukan posisi kotak dan ukurannya
kotak_posisi_x = 120
kotak_posisi_y = 130
kotak_lebar = 150
kotak_tinggi = 90

image_3[kotak_posisi_y:kotak_posisi_y + kotak_tinggi, kotak_posisi_x:kotak_posisi_x + kotak_lebar] = [255, 255, 255]

final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```
